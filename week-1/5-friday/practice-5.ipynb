```python
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f"{column}")
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
    print(f"Number of outliers: {outliers.shape[0]}")
    print("-" * 40)

    return outliers


age_outliers = find_outliers_iqr(titanic_data, "Age")
fare_outliers = find_outliers_iqr(titanic_data, "Fare")
```

**Кастомный imputer для подстановки по полу и пассажирскому классу:**

```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd


class AgeGroupMedianImputer(BaseEstimator, TransformerMixin):
    def __init__(self, age_col="Age", group_cols=["Pclass", "Sex"]):
        self.age_col = age_col
        self.group_cols = group_cols

    def fit(self, X, y=None):
        X = X.copy()

        self.group_medians_ = (
            X.groupby(self.group_cols)[self.age_col]
            .median()
        )

        self.global_median_ = X[self.age_col].median()

        return self

    def transform(self, X):
        X = X.copy()

        missing_age = X[self.age_col].isna()

        for idx in X[missing_age].index:
            group_key = tuple(X.loc[idx, self.group_cols])

            if group_key in self.group_medians_.index:
                X.loc[idx, self.age_col] = self.group_medians_.loc[group_key]
            else:
                X.loc[idx, self.age_col] = self.global_median_

        return X
    
```

**Предобработка данных: заполнение пропусков, масштабирование, кодирование**

Несколько трансформеров, и два отдельных препроцессора: один с кастомной подстановкой, один с подстановкой медианы через SimpleImpute().

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer

### SCALING & ENCODING & IMPUTING PIPELINE

age_col = ["Age"]
numerical_cols = ["SibSp", "Parch", "Fare"]
categorical_cols = ["Sex", "Embarked", "Title"]
pclass_col = ["Pclass"]


numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ('scaler', RobustScaler()) 
])

### оказывается он по умолчанию удаляет колонки, которых нет в column transformer !!!!
pclass_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent"))
])


categorical_transformer = Pipeline(steps =[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore')) 
])

age_scaler = Pipeline(steps =[
    ('scaler', RobustScaler()) 
])

age_median_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])


preprocessor_custom = ColumnTransformer(
    transformers= [
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols),
        ('age_scale', age_scaler, age_col),
        ("pclass", pclass_transformer, pclass_col)
    ]
)

preprocessor_median = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_cols),
        ("cat", categorical_transformer, categorical_cols),
        ("age", age_median_transformer, age_col),
        ("pclass", pclass_transformer, pclass_col)
    ]
)
```

**Пайплайны моделей**

Три пайплайна с подставлением `age` по медиане: Logistic Regression (`model_logreg_median`), Random Forest (`model_rf_median`) и Decision Tree (`model_dt_median`).

И также три пайплайна с кастомным подставлением: `model_logreg`, `model_rf` и `model_dt`.

```python
model_logreg = Pipeline(steps=[
    ("age_group_imputer", AgeGroupMedianImputer(
        age_col="Age",
        group_cols=["Pclass", "Sex"]
    )),
    ("preprocessor", preprocessor_custom),
    ("model", LogisticRegression(random_state=42, max_iter=1000))
])


model_dt = Pipeline(steps=[
    ("age_group_imputer", AgeGroupMedianImputer(
        age_col="Age",
        group_cols=["Pclass", "Sex"]
    )),
    ("preprocessor", preprocessor_custom),
    ("model", DecisionTreeClassifier(random_state=42))
])


model_rf = Pipeline(steps=[
    ("age_group_imputer", AgeGroupMedianImputer(
        age_col="Age",
        group_cols=["Pclass", "Sex"]
    )),
    ("preprocessor", preprocessor_custom),
    ("model", RandomForestClassifier(random_state=42))
])


model_logreg_median = Pipeline(steps=[
    ("preprocessor", preprocessor_median),
    ("model", LogisticRegression(random_state=42, max_iter=1000))
])


model_dt_median = Pipeline(steps=[
    ("preprocessor", preprocessor_median),
    ("model", DecisionTreeClassifier(random_state=42))
])


model_rf_median = Pipeline(steps=[
    ("preprocessor", preprocessor_median),
    ("model", RandomForestClassifier(random_state=42))
])
```

**Пространства поиска гиперпараметров**

```python
from scipy.stats import randint, loguniform
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold( ## потому что классификация и классы не сбалансированы
    n_splits=5,
    shuffle=True,
    random_state=42
)

# params

logreg_params = [
    {
        "model__solver": ["lbfgs"], # does not support l1 
        "model__penalty": ["l2"],
        "model__C": loguniform(0.001, 100), # sampling
        "model__max_iter": [1000]
    },
    {
        "model__solver": ["liblinear"],
        "model__penalty": ["l1", "l2"],
        "model__C": loguniform(0.001, 100),
        "model__max_iter": [1000]
    }
]


dt_params = {
    "model__criterion": ["gini", "entropy", "log_loss"],
    "model__max_depth": [None, 2, 3, 4, 5, 7, 10, 15],
    "model__min_samples_split": randint(2, 20),
    "model__min_samples_leaf": randint(1, 15),
    "model__max_features": [None, "sqrt", "log2"]
}


rf_params = {
    "model__n_estimators": randint(100, 700), # forest nufm
    "model__criterion": ["gini", "entropy", "log_loss"],
    "model__max_depth": [None, 3, 5, 7, 10, 15, 20],
    "model__min_samples_split": randint(2, 20),
    "model__min_samples_leaf": randint(1, 10),
    "model__max_features": [None, "sqrt", "log2"],
    "model__bootstrap": [True, False]
}
```

**Функция для поиска гиперпараметров через RandomizedSearchCV, валидации лучшей модели и вывода метрик**

```python
from sklearn.model_selection import RandomizedSearchCV

### find the best one

def search_model(
    model,
    param_distributions,
    model_name,
    X_train,
    y_train,
    X_valid,
    y_valid,
    n_iter=30 # then by default 450 fits in total for three models / 900 fits for 6 models
    # i chose 40 for all of them -> does 1200 fits = 40 * 5 * 6 
):

    print(f"\n{'='*60}")
    print(model_name)
    print(f"{'='*60}")

    search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_distributions,
        n_iter=n_iter,
        scoring={
            "accuracy": "accuracy",
            "f1": "f1"
        },
        refit="accuracy", # by this chooses which model to train on all data (not on splits from cv)
        cv=cv,
        random_state=42,
        n_jobs=-1, ## takes all my cpu
        verbose=1
    )

    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    preds = best_model.predict(X_valid)

    valid_accuracy = accuracy_score(y_valid, preds)
    valid_f1 = f1_score(y_valid, preds)

    print("\nBest CV Accuracy:", search.best_score_)
    print("Best parameters:")
    print(search.best_params_)

    print("\nValidation Accuracy:", valid_accuracy)
    print("Validation F1:", valid_f1)

    print("\nClassification Report:")
    print(classification_report(y_valid, preds))

    print("Confusion Matrix:")
    print(confusion_matrix(y_valid, preds))

    return {
    "model_name": model_name,
    "best_cv_accuracy": search.best_score_,
    "best_cv_f1": search.cv_results_["mean_test_f1"][search.best_index_],
    "valid_accuracy": valid_accuracy,
    "valid_f1": valid_f1,
    "best_params": search.best_params_,
    "best_model": best_model,
    "search_object": search,
    "preds": preds
}
```

**Запуск поиска гиперпараметров**

```python

from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

results_logreg = search_model(
    model=model_logreg,
    param_distributions=logreg_params,
    model_name="Logistic Regression by custom impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)

results_logreg_median = search_model(
    model=model_logreg_median,
    param_distributions=logreg_params,
    model_name="Logistic Regression by median impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)


results_dt = search_model(
    model=model_dt,
    param_distributions=dt_params,
    model_name="Decision Tree by custom impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)

results_dt_median = search_model(
    model=model_dt_median,
    param_distributions=dt_params,
    model_name="Decision Tree by median impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)


results_rf = search_model(
    model=model_rf,
    param_distributions=rf_params,
    model_name="Random Forest by custom impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)


results_rf_median = search_model(
    model=model_rf_median,
    param_distributions=rf_params,
    model_name="Random Forest by median impute",
    X_train=X_train,
    y_train=y_train,
    X_valid=X_valid,
    y_valid=y_valid,
    n_iter=40
)
```
